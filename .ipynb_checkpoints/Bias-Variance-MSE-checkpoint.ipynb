# 📘 Simulation Exercises: Bias–Variance–MSE in Ridge Regression

## 🧪 Exercise 1: Pointwise Bias², Variance, and MSE

**Goal:** Simulate multiple datasets, predict at a fixed point `x0`, and compute bias², variance, and MSE.

### Steps:

1. Define the true function:  
   $$ f(x) = 2x + \sin(1.5x) $$

2. Fix a test point:  
   $$ x_0 = 1.5 $$

3. For **B = 1000** simulations:
   - Sample **n = 50** training points:  
     $$ x_i \sim \text{Uniform}(-3, 3) $$
   - Generate responses:  
     $$ y_i = f(x_i) + \varepsilon, \quad \varepsilon \sim \mathcal{N}(0, 1) $$
   - Fit ridge regression with penalty parameter:  
     $$ \lambda = 1 $$
   - Predict:  
     $$ \hat{f}^{(b)}(x_0) $$

4. After all simulations, compute:
   - **Bias²**:  
     $$ \left( \frac{1}{B} \sum_{b=1}^B \hat{f}^{(b)}(x_0) - f(x_0) \right)^2 $$
   - **Variance**:  
     $$ \frac{1}{B} \sum_{b=1}^B \left( \hat{f}^{(b)}(x_0) - \bar{f}(x_0) \right)^2 $$
   - **MSE**:  
     $$ \frac{1}{B} \sum_{b=1}^B \left( \hat{f}^{(b)}(x_0) - f(x_0) \right)^2 $$

---

## 🧪 Exercise 2: Bias–Variance–MSE Averaged Over Multiple Points

**Goal:** Repeat the simulation but average the metrics across several test points.

### Steps:

1. Use the same function `f(x)` and data generation.

2. Create a test grid:  
   $$ x_{\text{test}} = [-2.5, -2.0, \dots, 2.5] $$  
   (11 evenly spaced points)

3. For **B = 500** simulations:
   - Fit ridge regression on new data.
   - Predict at each `x_test` point:  
     $$ \hat{f}^{(b)}(x_i) $$

4. For each test point, compute:
   - Bias²  
   - Variance  
   - MSE  
   (as in Exercise 1)

5. Average the results over all test points to get:
   - Average Bias²  
   - Average Variance  
   - Average MSE

---

## 🧪 Exercise 3: Bias–Variance–MSE vs. Regularization Strength

**Goal:** Analyze the bias–variance trade-off as a function of ridge penalty `λ`.

### Steps:

1. Use setup from Exercise 2.

2. Try different `λ` values:  
   $$ \lambda \in \{0, 0.1, 1, 10, 100\} $$

3. For each `λ`, repeat Exercise 2 and record:
   - Average Bias²  
   - Average Variance  
   - Average MSE

4. Plot:
   - `λ` on the x-axis (log scale)
   - Curves for Bias², Variance, and MSE on the y-axis
   - Optionally: add irreducible error line  
     $$ \sigma^2 = 1 $$

---


## Exercise 1

```python
from sklearn.linear_model import Ridge
import numpy as np

def f(x):
    return 2*x+np.sin(1.5*x)
def f_2(x):
    return 2*x-3*x**5
def f_3(x):
    return 2**x-x**2
B = 1000
y_pred = np.empty(B)
for b in range(0,B):
    rng = np.random.default_rng(b)
    x = rng.uniform(-3,3,50)
    epsilon = rng.normal(0,1,50)
    y = f(x) + epsilon
    model = Ridge(alpha=1.0)  # alpha = λ (regularization strength)
    model.fit(x.reshape(-1, 1),y)
    y_pred[b] = model.predict(np.array([[1.5]]))[0]
mean_pred = np.mean(y_pred)
bias_squared = (mean_pred - f(1.5))**2
variance = np.mean((y_pred-mean_pred)**2)
mse = np.mean((y_pred - f(1.5))**2)

print(f"Bias_squared: {bias_squared} - Variance: {variance} - MSE: {mse}")
```

## Exercise 2

```python
B=500
x_test =np.linspace(-2.5,2.5,11)
y_pred = np.empty((B,11))
for b in range(0,B):
    rng = np.random.default_rng(b)
    x = rng.uniform(-3,3,50)
    epsilon = rng.normal(0,1,50)
    y = f(x) + epsilon
    model = Ridge(alpha=1.0)  # alpha = λ (regularization strength)
    model.fit(x.reshape(-1, 1),y)
    y_pred[b] = model.predict(x_test.reshape(-1,1)).flatten()
mean_pred = np.mean(y_pred,axis=0)
bias_squared = (mean_pred-f(x_test))**2
var = np.var(y_pred,axis=0)
mse = np.mean((y_pred - f(x_test))**2,axis=0)
av_mse = np.mean(mse)
av_bias_squared = np.mean(bias_squared)
av_var = np.mean(var)
print(f"Mean prediction vector: {mean_pred}")
print(f"Squared Bias vector: {bias_squared}")
print(f"Variance vector: {var}")
print(f"MSE vector: {mse}")
print(f"Bias_squared: {av_bias_squared} - Variance: {av_var} - MSE: {av_mse}")
```

## Exercise 3

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import LogLocator, LogFormatterSciNotation

lamda = np.logspace(-6, 16, 23) 
B=500
x_test =np.linspace(-2.5,2.5,11)
X_test = np.power.outer(x_test,np.arange(10))

def bias_var_mse(B,X_test,lamda,n,f):
    y_pred = np.empty((B,len(lamda),len(x_test)))
    for b in range(0,B):
        rng = np.random.default_rng(b)
        x = rng.uniform(-3,3,n)
        X = np.power.outer(x,np.arange(10))
        epsilon = rng.normal(0,2,n)
        y = f(x) + epsilon
        for i,l in enumerate(lamda):
            model = Ridge(alpha=l)  # alpha = λ (regularization strength)
            model.fit(X,y)
            y_pred[b,i] = model.predict(X_test).flatten()
    mean_pred = np.mean(y_pred,axis=0)
    bias_squared = (mean_pred-f(x_test))**2
    var = np.var(y_pred,axis=0)
    mse = np.mean((y_pred - f(x_test))**2,axis=0)
    av_mse = np.mean(mse,axis=1)
    av_bias_squared = np.mean(bias_squared,axis=1)
    av_var = np.mean(var,axis=1)
    print(f"Bias_squared: {av_bias_squared} - Variance: {av_var} - MSE: {av_mse}")
    plt.plot(lamda, av_mse, label='mse', color='blue')
    plt.plot(lamda, av_bias_squared, label='bias squared', color='green')
    plt.plot(lamda, av_var, label='variance', color='orange')
    plt.xlabel('Lambda')
    plt.gca().xaxis.set_major_locator(LogLocator(base=10.0))
    plt.gca().xaxis.set_major_formatter(LogFormatterSciNotation())
    plt.xscale('log')
    plt.title(f)

    plt.legend()

    plt.show()
    
bias_var_mse(B,X_test,lamda,50,f)
bias_var_mse(B,X_test,lamda,50,f_2)
bias_var_mse(B,X_test,lamda,50,f_3)
```
